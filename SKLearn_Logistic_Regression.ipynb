Kita akan menggunakan logistic regression untuk memprediksi apakah seseorang akan membeli setelah melihat iklan sebuah produk. 

Steps:
1.  Ubah dataset menjadi Dataframe.
2.  Pisahkan atribut dan label.
3.  Hapus kolom 'User ID'.
4.  Latih model Logistic Regression.
5.  Evaluasi akurasi model.

```python
# Ubah dataset menjadi dataframe
import pandas as pd

df = pd.read_csv('/content/sample_data/Social_Network_Ads.csv')
df.head()
```

```python
df = df.dropna()
df.info()
```

```python
data = df.drop(columns=['User ID'])
data.head()
```

```python
# fungsi get_dummies() untuk melakukan proses One-Hot Encoding karena label pada dataset kita merupakan data kategorikal.

data = pd.get_dummies(data)
data
```

```python
# Pisahkan atribut dan label

X = data[["Age", "EstimatedSalary", "Gender_Female", "Gender_Male"]]
y = data["Purchased"]
```

```python
# lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

# latih model dengan fungsi fit
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Uji akurasi model

accuracy = model.score(X_test, y_test)
print('Accuracy: ', accuracy)
```
